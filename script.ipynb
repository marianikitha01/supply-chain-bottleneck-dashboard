```python
import pandas as pd
```

```python
# Loading the csv file
file_path = r"data/dynamic_supply_chain_logistics_dataset.csv"
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

```python
# Keep only relevant columns
columns_to_keep = [
    "timestamp", "vehicle_gps_latitude", "vehicle_gps_longitude",
    "fuel_consumption_rate", "eta_variation_hours",
    "traffic_congestion_level", "warehouse_inventory_level",
    "loading_unloading_time", "port_congestion_level",
    "supplier_reliability_score", "lead_time_days",
    "customs_clearance_time", "delay_probability",
    "risk_classification", "delivery_time_deviation"
]
df = df[columns_to_keep]
```

```python
# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
```

```python
# Create new derived columns
df["delay_flag"] = df["delivery_time_deviation"].apply(lambda x: "Delayed" if x > 0 else "On Time")
df["date"] = df["timestamp"].dt.date
```

```python
# Drop rows with missing timestamps
df = df.dropna(subset=["timestamp"])
```

```python
# Save the cleaned file
df.to_csv("data/cleaned_supply_chain_dashboard.csv", index=False)

print("Cleaned data saved ")
```
